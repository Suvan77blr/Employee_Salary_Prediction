```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Initial Loading

```python
dataset_path = "/content/drive/MyDrive/3rd_Yr/IBM-Edunet_Internship/Numerical_Done_DF.csv"
```

```python
# Loading the given dataset.
# df = pd.read_csv(dataset_path, na_values="?")
df = pd.read_csv(dataset_path)
# df = pd.read_csv(r"./emp_data.csv", na_values="?")
```

```python
df.head()
```

### Encoding

##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
categorical_cols.remove('income')
print(f"Columns to encode: {categorical_cols}")
```

```python
# Label Encoding each of the categorical columns.
for col in categorical_cols:
  df[col] = encoder.fit_transform(df[col])
```

```python
df.head()
```

##### Info

First splitting the data, then applying scaling to it.


##### `I/O` & `Train-Test` Split

```python
x = df.drop(columns=['income'])
y = df['income']
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train-Test split => then applying scaling.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

##### MinMax-Scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled
```

### Applying ML Algos

##### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train_scaled, y_train)
# knn.fit(x_train, y_train)
```

```python
predict = knn.predict(x_test_scaled)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predict)
```

```python
knn2 = KNeighborsClassifier(n_neighbors=10)
knn2.fit(x_train_scaled, y_train)
predict2 = knn2.predict(x_test_scaled)
```

```python
accuracy_score(y_test, predict2)
```

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi_reg = LogisticRegression()
logi_reg.fit(x_train_scaled, y_train)
logi_predict = logi_reg.predict(x_test_scaled)
```

```python
accuracy_score( y_test, logi_predict)
```

##### Neural Network => MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clf_1 = MLPClassifier(
    solver='lbfgs',
    hidden_layer_sizes=(7, 4),
    random_state=3,
    max_iter=2000
)
mlp_clf_1.fit(x_train_scaled, y_train)
# => 0.8487819983484723
```

```python
# With (7, 4)
mlp_clf_predict_1 = mlp_clf_1.predict(x_test_scaled)
accuracy_score( y_test, mlp_clf_predict_1)
```

```python
# Alternative Next-best
mlp_clf2 = MLPClassifier(
    solver='lbfgs',
    hidden_layer_sizes=(5, 2),
    random_state=3,
    max_iter=2500
)
mlp_clf2.fit(x_train_scaled, y_train)
```

```python
# => lbfgs w/ 2500.
mlp_clf_predict2 = mlp_clf2.predict(x_test_scaled)
accuracy_score( y_test, mlp_clf_predict2)
```
